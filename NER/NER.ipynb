```python
product = []
for t in txt[:10]:
    s = Sentence(t)
    tagger1.predict(s)
    
    p = []
    if len(s.get_labels()) > 0:
        for e in s.get_labels():
            if e.value == 'PRODUCT':
                p.append(e.data_point.text.replace('the ', ''))
    
    product.append(p)
    
product
```

```python
phrase = []
for t in txt[:10]:
    s = Sentence(t)
    tagger2.predict(s)
    
    p = []
    if len(s.get_labels()) > 0:
        for e in s.get_labels():
            if e.value == 'NP':
                p.append(e.data_point.text)
    
    phrase.append(p)

phrase
```

```python
f = open(r'food_vocab.txt','r')
food = list(f)
f.close()
food = [f.rstrip('\n') for f in food]
food
```

```python
def clean_phrase(p):
    stop_words = ['The ', 'A ', 'An ', 'My ', 'Favorite ', 'Great ', 'Good ', 'Incredible ', 'Amazing ', 'Really ', 'Nice ']
    for s in stop_words:
        p = p.replace(s, '')
        p = p.replace(s.lower(), '')
        p = p.strip('\"')
    return p
```

```python
food_phrase = []
for p_list in phrase:
    p_food = []
    for p in p_list:
        p_ = p.split(' ')
        if len(list( set(p_) & set(food)) ) > 0:
            p_food.append(clean_phrase(p))
    food_phrase.append(p_food)
        
food_phrase
```
