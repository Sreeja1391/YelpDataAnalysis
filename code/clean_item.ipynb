#### clean products

```python
products = []
for w in df['product'].tolist():
    products += w
    
c = Counter(products)

freq = list(c.values())
n_invalid = sum([f<=10 for f in freq]) 
invalid = c.most_common()[-n_invalid:-1]
invalid = [i_n[0] for i_n in invalid]

len(invalid)
```

```python
invalid_ = ['the', 'The', 'iPad', 'iPhone', 'Quick Fixx', 'Uber', 'OJ', 'pad see ew', 'Inch of Heaven', 'Tom', 'mi', 'OG', 
            'Funky Monkey', 'Halo Halo', 'Ringer', 'Lumberjack', 'Painkiller', 'Mugshots', 'Dirty Jersey', 'NoWait', 'Wawa',
            'DoorDash', 'Stuffed', 'Pho', 'pad kee mow', 'Maverick', 'BLT', 'PBR', 'Yelp', 'Fuel', 'KFT', 'Barking Chihuahua',
            'Thai', 'Groupon', 'Done and Dusted', 'Tots']
non_vegan = ['Duck Kebab', 'Crab Rangoon', 'Penicillin', 'Chicken Cutlet', 'Caviar', 'Black Pig', 'Cheesesteak', 'Butter Chicken', 
             'Chicken TikkMasala', 'Huevos Rancheros', 'Philly Steak',
             'Eggs Benedict', 'eggs Benedict', 'Eggs Benedict Florentine',
             'Philly Cheesesteak', 'Philly Cheese Steak',
             'WizKid Philly', 'Wiz Kid Philly', 'Philly', 'Wizkid', 'Wiz Kid', 'WK Philly',
             'Ziggy Burger', 'Ziggy']
invalid = invalid + invalid_ + non_vegan
```

```python
rep = [['DDNoodles', 'Spicy DDNoodles', 'Ddnoodles', 'ddnoodles', 'DDnoodles', ],
       ['Udon', 'Udon Noodle'],
       ['Bloody Mary', 'Bloody Marys', "Bloody Mary's"], 
       [ 'Spring Roll', 'Spring Rolls'],
       ['Mac&Cheese', 'Mac', 'Mac n cheese', 'Mac and cheese', 'Mac and Cheese'],
       ['Coke', 'coke'],
       ['DDan', 'dan']]
```

```python
product = df['product'].tolist()
product_ = []
for p in product:
    p_ = p
    if len(p) > 0:
        p_ = list((set(p_) - set(invalid)))
        for r in rep:
            if len(list(set(p) & set(r))) > 0:
                p_ = list((set(p) - set(r)))
                p_.append(r[0])
    product_.append(p_)
```

```python
product_
```

```python
products = []
for w in product_:
    products += w
    
c_ = Counter(products)
freq = list(c_.values())
n_invalid_ = sum([f<=10 for f in freq])

c_.most_common()[:-n_invalid_]
```

```python
common_product = c_.most_common()[:-n_invalid_]
len(common_product)
```
