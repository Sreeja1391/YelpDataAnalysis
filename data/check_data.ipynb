```python
import pandas as pd

df = pd.read_json('/afs/cs.wisc.edu/u/l/e/leng/public/yelp_Fall2023/business.json', lines = True)
df
```

```python
categ = df['categories'].tolist()
```

```python
categ
```

```python
restaurant = []
for c in categ:
    print(c)
    if c != None and 'Restaurants'in c:
        restaurant.append(1)
    else:
        restaurant.append(0)
        
sum(restaurant)
```

```python
sum(restaurant)
```

```python
df['restaurant'] = restaurant
rest_df = df[df['restaurant'] == 1]
rest_df
```

```python
attri_ = rest_df['attributes'].tolist()
categ_ = rest_df['categories'].tolist()
```

```python
rest_attri = []

for i in range(rest_df.shape[0]):
    if attri_[i] == None:
        continue
    rest_attri = list(set(rest_attri) | set(list(attri_[i].keys())))
```

```python
rest_attri
```

```python
diet_restrict = []
for i in range(rest_df.shape[0]):
    restrict = 0
    if attri_[i] != None:
        if 'DietaryRestrictions' in list(attri_[i].keys()):
            restrict = 1
    
    diet_restrict.append(restrict)
```

```python
sum(diet_restrict)
```

```python
rest_df['diet_restrict'] = diet_restrict
df1 = rest_df[rest_df['diet_restrict'] == 1]
```

```python
for a in df1['attributes'].tolist():
    print(a['DietaryRestrictions'])
```

```python
rest_categ = []

for i in range(rest_df.shape[0]):
    if categ_[i] == None:
        continue
    rest_categ = list(set(rest_categ) | set(categ_[i].replace(' ', '').split(',')))
```

```python
rest_categ.sort()
rest_categ
```

```python
categ_vegan = []
for i in range(rest_df.shape[0]):
    vegan = 0
    if categ_[i] != None:
        c = categ_[i].replace(' ', '').split(',')
        if 'Vegan' in c or 'Vegetarian' in c:
             vegan = 1
    
    categ_vegan.append(vegan)
    
sum(categ_vegan)
```

```python
rest_df['categ_vegan'] = categ_vegan
rest_vegan_df = rest_df[rest_df['categ_vegan'] == 1]
rest_vegan_df
```
